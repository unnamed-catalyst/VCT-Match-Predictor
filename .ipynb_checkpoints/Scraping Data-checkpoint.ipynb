```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from webdriver_manager.chrome import ChromeDriverManager

# Setup Selenium
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

try:
    # Navigate to the page
    url = "https://www.rib.gg/series/100-thieves-vs-nrg-esports-champions-tour-2025-americas-kickoff/83724?tab=team-stats"
    driver.get(url)
    
    # Give the page time to load (fixed wait instead of conditional)
    print("Waiting for page to load...")
    time.sleep(10)  # Wait 10 seconds for content to load
    
    # Get all the stat rows using a more general selector
    print("Extracting stats...")
    stats = {}
    
    # Try different selectors that might contain the stats
    possible_selectors = [
        "div[class*='css-pr7wk6']",
        "div[class*='flex']",
        "div.row",  # Generic row class
        "div[style*='display: flex']"  # Style-based selector
    ]
    
    for selector in possible_selectors:
        stat_rows = driver.find_elements(By.CSS_SELECTOR, selector)
        print(f"Found {len(stat_rows)} elements with selector: {selector}")
        
        for row in stat_rows:
            # Try to identify if this is a stat row by checking for 3 child divs
            divs = row.find_elements(By.TAG_NAME, "div")
            if len(divs) == 3:
                left_text = divs[0].text.strip()
                middle_text = divs[1].text.strip()
                right_text = divs[2].text.strip()
                
                # Only process rows that look like stat rows (middle has text, sides have numbers)
                if middle_text and (left_text.isdigit() or right_text.isdigit() or "%" in left_text or "%" in right_text):
                    stats[middle_text] = {"100 Thieves": left_text, "NRG Esports": right_text}
                    print(f"{middle_text}: {left_text} - {right_text}")
    
    # Display pistol-related stats
    print("\nExtracted Stats:")
    for stat_name, values in stats.items():
        if "Pistol" in stat_name:
            print(f"\nFound Pistol Stat: {stat_name}")
            print(f"100 Thieves: {values['100 Thieves']}")
            print(f"NRG Esports: {values['NRG Esports']}")

except Exception as e:
    print(f"Error occurred: {e}")

finally:
    # Always close the browser
    driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup
import re

# Function to check if a string might be a stat value
def is_likely_stat_value(text):
    return bool(re.match(r'^\d+(\.\d+)?%?$', text.strip()))

# Get the page
url = "https://www.rib.gg/series/100-thieves-vs-nrg-esports-champions-tour-2025-americas-kickoff/83724?tab=team-stats"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

print("Fetching the page...")
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

# Try to find divs containing pistol-related text
print("Searching for pistol stats...")
pistol_divs = soup.find_all(string=re.compile("BREAKDOWN", re.IGNORECASE))

for div in pistol_divs:
    print(f"Found text: {div}")
    parent = div.parent
    
    # Try to find the surrounding elements
    while parent and parent.name != "div":
        parent = parent.parent
    
    # Look at siblings of this element
    if parent:
        prev_sibling = parent.find_previous_sibling()
        next_sibling = parent.find_next_sibling()
        
        if prev_sibling and next_sibling:
            print(f"Stat: {div}")
            print(f"Left value: {prev_sibling.text.strip()}")
            print(f"Right value: {next_sibling.text.strip()}")
```

```python
import requests
import json

# API endpoint you provided
api_url = "https://be-prod.rib.gg/v1/matches/188335/series"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(api_url, headers=headers)
data = response.json()

# Pretty print the first level of the JSON structure to see what's available
print(json.dumps({k: type(v).__name__ for k, v in data.items()}, indent=2))

# Look for anything related to pistols or stats
pistol_stats = {}

def search_dict_for_pistol(d, path=""):
    for k, v in d.items():
        new_path = f"{path}.{k}" if path else k
        if isinstance(v, dict):
            search_dict_for_pistol(v, new_path)
        elif isinstance(v, list):
            for i, item in enumerate(v):
                if isinstance(item, dict):
                    search_dict_for_pistol(item, f"{new_path}[{i}]")
        elif isinstance(v, (str, int, float)) and "pistol" in str(k).lower():
            print(f"Found pistol-related key: {new_path} = {v}")
            pistol_stats[new_path] = v

search_dict_for_pistol(data)
```

```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Setup Chrome options
firefox_options = Options()
firefox_options.add_argument("--headless")  # Run in headless mode
firefox_options.add_argument("--no-sandbox")
firefox_options.add_argument("--disable-dev-shm-usage")

# Initialize the driver
driver = webdriver.Firefox(options=firefox_options)

# Navigate to the page
url = "https://www.rib.gg/series/100-thieves-vs-nrg-esports-champions-tour-2025-americas-kickoff/83724?tab=team-stats"
driver.get(url)

# Wait for the page to load
wait = WebDriverWait(driver, 20)
wait.until(EC.presence_of_element_located((By.XPATH, "//div[contains(text(), 'Breakdown')]")))

# Wait additional time for all data to load
time.sleep(5)

# Find the breakdown section
breakdown_section = driver.find_element(By.XPATH, "//div[contains(text(), 'Breakdown')]/ancestor::div[contains(@class, 'MuiBox-root')]")

# Extract the table data
rows = breakdown_section.find_elements(By.XPATH, ".//div[contains(@class, 'MuiBox-root') and @role='row']")

# Create lists to store the data
stats = []
team1_values = []
team2_values = []

# Extract data from each row
for row in rows:
    # Skip the team names row (first row)
    cells = row.find_elements(By.XPATH, "./div")
    if len(cells) == 3:  # Only process rows with 3 columns
        stat_name = cells[1].text.strip()
        if stat_name:  # Only process non-empty stat names
            team1_value = cells[0].text.strip()
            team2_value = cells[2].text.strip()
            
            stats.append(stat_name)
            team1_values.append(team1_value)
            team2_values.append(team2_value)

# Create a DataFrame
df = pd.DataFrame({
    '100 Thieves': team1_values,
    'Stat': stats,
    'NRG Esports': team2_values
})

# Save to CSV
# df.to_csv('match_stats.csv', index=False)

# Print the DataFrame
print(df)

# Close the driver
driver.quit()
```

```python
driver = uc.Chrome()
driver.get("https://www.rib.gg/series/100-thieves-vs-nrg-esports-champions-tour-2025-americas-kickoff/83724?tab=team-stats")

print("Waiting for page to load...")
time.sleep(10)  # Give the page time to fully load

# Now let's find the stat containers
# First, let's try to locate the team names to identify the structure
team_names = driver.find_elements(By.XPATH, "//div[contains(@class, 'css-') and contains(text(), 'Thieves') or contains(text(), 'NRG')]")
print(f"Found {len(team_names)} team name elements")
if team_names:
    for team in team_names:
        print(f"Team found: {team.text}")

# Let's look for the stat categories and their containers
print("\nExtracting stat categories and values...")

# Try to find the parent containers for the stats
stat_containers = driver.find_elements(By.XPATH, "//div[contains(@class, 'css-') and ./div[contains(text(), 'Pistol Rounds Won') or contains(text(), 'First Kills') or contains(text(), 'KAST')]]")
print(f"Found {len(stat_containers)} stat container elements")

# If we can't find containers, let's extract data directly around the stat names
all_stats = {}
stat_names = driver.find_elements(By.XPATH, "//div[contains(text(), 'Pistol Rounds Won') or contains(text(), 'First Kills') or contains(text(), 'KAST') or contains(text(), 'K:D') or contains(text(), 'ADR') or contains(text(), 'ACS')]")

for stat in stat_names:
    stat_name = stat.text
    print(f"\nFound stat: {stat_name}")
    
    # Look for the closest parent containing both the stat name and team values
    parent = stat.find_element(By.XPATH, "./ancestor::div[contains(@class, 'css-')][position()=3]")
    
    # Find all value elements within this parent
    # This XPath is complex but tries to find only direct value elements, not nested stats
    value_elements = parent.find_elements(By.XPATH, ".//div[contains(@class, 'css-') and string-length(text()) > 0 and string-length(text()) < 10 and not(contains(text(), 'Pistol')) and not(contains(text(), 'First')) and not(contains(text(), 'KAST')) and not(contains(text(), 'K:D')) and not(contains(text(), 'ADR')) and not(contains(text(), 'ACS'))]")
    
    values = [el.text for el in value_elements if el.text.strip()]
    
    if len(values) >= 2:  # We expect at least 2 values (one for each team)
        print(f"  Values: {values}")
        all_stats[stat_name] = values

# Create a DataFrame from the collected data
if all_stats:
    # Assuming first team is 100 Thieves and second is NRG based on URL
    teams = ["100 Thieves", "NRG Esports"]
    
    # Transform data for DataFrame
    data = {}
    for stat_name, values in all_stats.items():
        if len(values) >= 2:
            for i, team in enumerate(teams):
                if i < len(values):
                    if team not in data:
                        data[team] = {}
                    data[team][stat_name] = values[i]
    
    # Create DataFrame
    df = pd.DataFrame(data)
    print("\nExtracted Data:")
    print(df)
    
    # Save to CSV
    df.to_csv("team_stats.csv")
    print("Data saved to team_stats.csv")
else:
    print("Failed to extract meaningful stat data")

driver.quit()
```

### Pistol Rounds, First Kills, Clutch%

```python
driver = uc.Chrome()
driver.get("https://www.rib.gg/series/100-thieves-vs-nrg-esports-champions-tour-2025-americas-kickoff/83724?tab=team-stats")

print("Waiting for page to load...")
time.sleep(10)  # Give the page time to fully load

# Extract the team names
team_elements = driver.find_elements(By.XPATH, "//p[contains(@class, 'css-1aujtoy') and string-length(text()) > 0]")
teams = [team.text.strip() for team in team_elements if len(team.text.strip()) > 0][:2]  # Extract first two team names

print(f"Teams detected: {teams}")

result_elements = driver.find_elements(By.XPATH, "//p[contains(@class, 'css-tesx4r') or contains(@class, 'css-hov9j3') and string-length(text()) > 0]")
results = [result.text.strip() for result in result_elements if len(result.text.strip()) > 0][:2]  # Extract the win and loss results

print(f"Results detected: {results}")

# Extract the statistics
all_stats = {}

# Find all stat categories
stat_names = driver.find_elements(By.XPATH, "//div[contains(text(), 'Pistol Rounds Won') or contains(text(), 'First Kills') or contains(text(), 'KAST') or contains(text(), 'Clutches')]")

for stat in stat_names:
    stat_name = stat.text
    print(f"\nFound stat: {stat_name}")

    try:
        # Look for numerical values in adjacent elements
        team1_value_elements = stat.find_elements(By.XPATH, "./preceding-sibling::div[contains(@class, 'css-') and string-length(text()) > 0]")

#         team1_values = [el.text.strip() for el in team1_value_elements if el.text.strip().replace('%', '').replace('.', '').isdigit()]  # Ensure values are numbers

        
        team2_value_elements = stat.find_elements(By.XPATH, "./following-sibling::div[contains(@class, 'css-') and string-length(text()) > 0]")

#         team2_values = [el.text.strip() for el in team2_value_elements if el.text.strip().replace('%', '').replace('.', '').isdigit()]  # Ensure values are numbers
        
        
        print(f"  {teams[0]} {stat_name}: {team1_value_elements[0].text.strip().replace('%', '').replace('.', '')}")
        print(f"  {teams[1]} {stat_name}: {team2_value_elements[0].text.strip().replace('%', '').replace('.', '')}")
#         if len(values) == 2:  # We expect exactly two values (one per team)
#             all_stats[stat_name] = values
#         else:
#             print(f"  Skipped due to unexpected format: {values}")

        all_stats[stat_name] = [team1_value_elements[0].text.strip().replace('%', '').replace('.', ''), team2_value_elements[0].text.strip().replace('%', '').replace('.', '')]
    
    except Exception as e:
        print(f"Error extracting {stat_name}: {e}")

# Create a DataFrame
if all_stats and len(teams) == 2:
    main_df = pd.DataFrame.from_dict(all_stats, orient='index', columns=teams).transpose()
    print("\nExtracted Data:")
    print(main_df)

    # Converting the clutches into percentages
    main_df["Clutches"] = main_df["Clutches"].str.split(" / ").apply(lambda x: int(x[0]) / int(x[1]) if int(x[1]) != 0 else 0)
    main_df['KAST'] = main_df['KAST'].astype('float').div(100)
    
    main_df = main_df.astype('float')
    
    # Save to CSV
#     df.to_csv("team_stats.csv")
#     print("Data saved to team_stats.csv")
else:
    print("Failed to extract meaningful stat data")

driver.quit()
```

```python
# Convert "X / Y" to percentage
def to_percentage(value):
    if (value == '-'):
        return f"{(0 / 1) :.2f}"
    x, y = map(int, value.split(' / '))
    return f"{(x / y) :.2f}"

# Mapping function
def get_advantage(xvy):
    x, y = map(int, xvy.split('v'))
    return f"+{x - y}" if x - y > 0 else f"{x - y}"
```

```python
driver = uc.Chrome()
driver.get("https://www.rib.gg/series/100-thieves-vs-nrg-esports-champions-tour-2025-americas-kickoff/83724?tab=team-stats")

print("Waiting for page to load...")
time.sleep(10)  # Give the page time to fully load

# Extract the team names
team_elements = driver.find_elements(By.XPATH, "//p[contains(@class, 'css-1aujtoy') and string-length(text()) > 0]")
teams = [team.text.strip() for team in team_elements if len(team.text.strip()) > 0][:2]  # Extract first two team names

print(f"Teams detected: {teams}")

result_elements = driver.find_elements(By.XPATH, "//p[contains(@class, 'css-tesx4r') or contains(@class, 'css-hov9j3') and string-length(text()) > 0]")
results = [result.text.strip() for result in result_elements if len(result.text.strip()) > 0][:2]  # Extract the win and loss results

print(f"Results detected: {results}")

# Extract the statistics
all_stats = {}

# Find all stat categories
stat_names = driver.find_elements(By.XPATH, "//div[contains(text(), '5v4') or contains(text(), '4v5') or contains(text(), '4v4') or contains(text(), '4v3') or contains(text(), '3v4') or contains(text(), '3v3') or contains(text(), '5v3') or contains(text(), '3v5') or contains(text(), '3v2') or contains(text(), '2v3') or contains(text(), '4v2') or contains(text(), '2v4') or contains(text(), '2v2') or contains(text(), '3v1') or contains(text(), '1v3') or contains(text(), '1v1') or contains(text(), '2v1') or contains(text(), '1v2') or contains(text(), '1v4') or contains(text(), '4v1') or contains(text(), '5v2') or contains(text(), '2v5') or contains(text(), '5v1') or contains(text(), '1v5')]")

for stat in stat_names:
    stat_name = stat.text
    print(f"\nFound stat: {stat_name}")

    try:
        # Look for numerical values in adjacent elements
        team1_value_elements = stat.find_elements(By.XPATH, "./parent::div/parent::div/preceding-sibling::div[contains(@class, 'css-') and string-length(text()) > 0]")

#         team1_values = [el.text.strip() for el in team1_value_elements if el.text.strip().replace('%', '').replace('.', '').isdigit()]  # Ensure values are numbers

        
        team2_value_elements = stat.find_elements(By.XPATH, "./parent::div/parent::div/following-sibling::div[contains(@class, 'css-') and string-length(text()) > 0]")

#         team2_values = [el.text.strip() for el in team2_value_elements if el.text.strip().replace('%', '').replace('.', '').isdigit()]  # Ensure values are numbers
        
        
        print(f"  {teams[0]} {stat_name}: {team1_value_elements[0].text.strip()}")
        print(f"  {teams[1]} {stat_name}: {team2_value_elements[0].text.strip()}")
#         if len(values) == 2:  # We expect exactly two values (one per team)
#             all_stats[stat_name] = values
#         else:
#             print(f"  Skipped due to unexpected format: {values}")

        all_stats[stat_name] = [team1_value_elements[0].text.strip().replace('%', '').replace('.', ''), team2_value_elements[0].text.strip().replace('%', '').replace('.', '')]
    
    except Exception as e:
        print(f"Error extracting {stat_name}: {e}")

# Create a DataFrame
if all_stats and len(teams) == 2:
    xvy_df = pd.DataFrame.from_dict(all_stats, orient='index', columns=teams).transpose()
    print("\nExtracted Data:")
#     print(df)

    # Apply transformation to all columns except 'Team'
    for col in xvy_df.columns:  
        xvy_df[col] = xvy_df[col].apply(to_percentage)

    xvy_df = xvy_df.astype('float')
    
    # Create new aggregated DataFrame
    agg_data = {}
    counts = {}

    for col in xvy_df.columns:
        category = get_advantage(col)
        if category not in agg_data:
            agg_data[category] = xvy_df[col].copy()
            counts[category] = 1
        else:
            agg_data[category] += xvy_df[col]
            counts[category] += 1

    # Convert to DataFrame and take the average
    xvy_df = pd.DataFrame({key: agg_data[key] / counts[key] for key in agg_data})

    # Save to CSV
#     df.to_csv("team_stats.csv")
#     print("Data saved to team_stats.csv")
else:
    print("Failed to extract meaningful stat data")

driver.quit()
```
